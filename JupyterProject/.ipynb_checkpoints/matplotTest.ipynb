```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_axis=[1,2,3]
y_axis=[4,5,6]
plt.plot(x_axis,y_axis)
plt.title("My First Graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
x1axis=[2,4,6, 8]
y1axis=[1,10,100,1000]
x2axis=[1,3,5,7]
y2axis=[100,110,120,130]
```

```python
plt.title("Two Plots on One Chart")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x1axis,y1axis,'r')
plt.plot(x2axis,y2axis,'g')
plt.legend('exponential','linear')
```
